```python
# function to find upper and lower limits using inter quartile range(IQR)

def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
lr,ur=remove_outlier(df['rate_marriage'])
lr,ur
```

```python
# capping outliers in 'rate_marriage','children' and 'occupation' columns using upper and lower limits
```

```python
df['rate_marriage']=np.where(df['rate_marriage']>ur,ur,df['rate_marriage'])
df['rate_marriage']=np.where(df['rate_marriage']<lr,lr,df['rate_marriage'])
```

```python
lr,ur=remove_outlier(df['children'])
lr,ur
```

```python
df['children']=np.where(df['children']>ur,ur,df['children'])
df['children']=np.where(df['children']<lr,lr,df['children'])
```

```python
lr,ur=remove_outlier(df['occupation'])
lr,ur
```

```python
df['occupation']=np.where(df['occupation']>ur,ur,df['occupation'])
df['occupation']=np.where(df['occupation']<lr,lr,df['occupation'])
```

```python
# separating features and Target to X and y respectively
```

```python
X = df.drop('affairs', axis=1)
y = df['affairs']
```

```python
X.columns
```

```python
# Using train_test_split to split data to train and test dataset
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
# Fitting our model
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# prediction & evalaution
```

```python
y_pred = model.predict(X_test)
```

```python
model_score= model.score(X_test,y_test)
model_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
